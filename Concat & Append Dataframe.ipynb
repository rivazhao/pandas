```python
import pandas as pd
import numpy as np 
from IPython.display import display
from IPython.display import Image
```

##Create a dataframe

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a
```

##Create a second dataframe

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

##create a third dataframe for further explaination 

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_n = pd.DataFrame(raw_data, columns = ['subject_id','test_id'])
df_n
```

##Join two dataframe along the rows

```python
df_new=pd.concat([df_a, df_b])
df_new
```

```python
##Join two dataframe along the column
```

```python
df_new_column=pd.concat([df_a, df_b],axis=1)
df_new_column
```

##Join two dataframe along the rows using append 
A useful shortcut to concat are the append instance methods on Series and DataFrame.
These methods actually predated concat. They concatenate along axis=0, namely the index

```python
df_new2=df_a.append(df_b)
df_new2
```
